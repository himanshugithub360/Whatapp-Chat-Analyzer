```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Apne Group 😎.txt','r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
data
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2},\s+\d{1,2}:\d{2}\s*[ap]m\s-\s'
```

```python
messages = re.split(pattern, data, flags=re.IGNORECASE)[1:]
messages
```

```python
import re

messages = [
    re.sub(r'[\u2066-\u2069]', '', msg).replace('@', ' ').strip()
    for msg in messages
]
```

```python
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
dates = [d.replace('\u202f', ' ') for d in dates]
```

```python
dates
```

```python
import pandas as pd

df = pd.DataFrame({
    'user_message': messages,
    'message_date': dates
})

# Replace Unicode narrow space with a normal space


# Convert to datetime
df['message_date'] = pd.to_datetime(
    df['message_date'],
    format='%d/%m/%y, %I:%M %p - '
)

df['date_12hr'] = df['message_date'].dt.strftime('%d/%m/%Y, %I:%M %p').str.lower()

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
users = []
messages = []

for msg in df['user_message']:
    if ': ' in msg:
        user, message = msg.split(': ', 1)  # Split only once
        users.append(user)
        messages.append(message)
    else:
        users.append('group_notification')
        messages.append(msg)

df['user'] = users
df['message'] = messages
```

```python
df.drop(columns=['user_message'], inplace=True)
```
